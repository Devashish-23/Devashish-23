```
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Load CSV (upload file in Colab first)
df = pd.read_csv("/content/Bhagwad_Gita.csv.csv")

# Fill missing values
df = df.fillna("")

# Check columns
print(df.columns)
df.head()
```

```
chapter_counts = df['Chapter'].value_counts().sort_index()

plt.figure()
plt.bar(chapter_counts.index, chapter_counts.values)
plt.xlabel("Chapter")
plt.ylabel("Number of Verses")
plt.title("Number of Verses per Chapter (Bhagavad Gita)")
plt.show()
```

```
df['shloka_length'] = df['Shloka'].apply(lambda x: len(x.split()))

plt.figure()
plt.hist(df['shloka_length'], bins=30)
plt.xlabel("Words per Shloka")
plt.ylabel("Frequency")
plt.title("Distribution of Sanskrit Shloka Lengths")
plt.show()
```

```
def clean_hindi_text(text):
    text = text.lower()
    text = re.sub(r'[^\u0900-\u097F\s]', '', text)  # keep Devanagari
    return text

all_hindi_text = " ".join(df['HinMeaning'].tolist())
words = clean_hindi_text(all_hindi_text).split()

word_freq = Counter(words)
top_words = word_freq.most_common(20)

words, counts = zip(*top_words)

plt.figure()
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.xlabel("Hindi Words")
plt.ylabel("Frequency")
plt.title("Top 20 Most Frequent Hindi Words")
plt.show()
```

```
def detect_language(text):
    if re.search(r'[अ-ह]', text):
        return "Devanagari Script"
    else:
        return "Other"

df['Language'] = df['Shloka'].apply(detect_language)

lang_counts = df['Language'].value_counts()

plt.figure()
plt.bar(lang_counts.index, lang_counts.values)
plt.xlabel("Language Type")
plt.ylabel("Count")
plt.title("Language Distribution (Script-Based)")
plt.show()
```
